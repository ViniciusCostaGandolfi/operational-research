# Opencast Mining

Maximize $\sum_{j=0}^n$ $Lucro_{j}$ * $x_{j}$ 

$blocks_{LevelI} ≤ blocks_{LevelJ} ∀ Level_{i, j}$

n = 30

m = 30

$i = 1, 2, 3, .. n$

$j = 1, 2, 3, .. m$

$x_j$ = {0, 1} ∀ j = 1... n

```python
# Inicializa o problema p que será resolvido com o solver 
model = ConcreteModel()
```

```python
# Fazendo os Sets
model.j = Set(initialize=j)
```

```python
model.x = Var(model.j, domain=Binary, bounds=(0, 1))
```

```python
# Declarando a função objetivo
expr = sum(model.x[j] * blocks[j] for j in model.j)
model.obj = Objective(sense=maximize, expr=expr)
```

```python
# Criando as restrições
model.rest1 = ConstraintList()
for i, j in levels:
  b2 = model.x[j]
  b1 = model.x[i]
  model.rest1.add(b2 >= b1)
```

```python
model.pprint()
```

```python
solver = SolverFactory('cbc', executable='/usr/bin/cbc')
results = solver.solve(model, tee=True)
```

```python
df = pd.DataFrame(columns=["X", "Preço", "Status"], index=np.arange(1, 31))

for j in model.j:
  df['X'][j]      = j
  df["Preço"][j]  = blocks[j]
  df["Status"][j] = value(model.x[j])

print(f"Valor Máximo da função objetivo: {model.obj()}")
df
```
