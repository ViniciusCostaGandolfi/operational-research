# Opencast Mining

Maximizar: $\sum_{i=1}^m$ $lucro_{i}$ * $x_{i}$ 

$x_{i} ≥ x_{j} ∀  associação_{i, j}$

$i = 1, 2, 3, .. m$

$j = 1, 2, 3, .. n$

m = 30; número de linhas

n = 30; número de colunas


$x_i$ = {0, 1}

```python
J = np.arange(1, 31) # contagem de 1 até 30
I = np.arange(1, 31) # contagem de 1 até 30

print(f"""
I = {I}
J = {J}
""")
```

```python
# Inicializa o problema como MIP e usando o solver SCIP para MIP
solver = pywraplp.Solver("Opencast Mining", pywraplp.Solver.SCIP_MIXED_INTEGER_PROGRAMMING)
```

```python
# Definindo as variáveis binárias
x = {}
for j in J:
  x[j] = solver.IntVar(0, 1, f"x[{j}]")
  print(x[j])
```

# Maximizar $\sum_{j=0}^n$ $Lucro_{j}$ * $x_{j}$ 

```python
# Definindo a função objetivo
obj = []

for j in J:
  obj.append(Lucro[j] * x[j])
  if j != 30:
    print(f" x[{j}]*{Lucro[j]} +", end="")
  else:
    print(f" x[{j}]*{Lucro[j]}")
```

```python
# Definindo as restrtições de Level
c = 0
for i, j in levels:
  c+=1
  b1 = x[i]
  b2 = x[j]
  solver.Add(b1 <= b2)
  print(f"Restrição {c}: x[{i}] ≥ x[{j}]")
```

```python
# Resolvendo o problema
solver.Maximize(solver.Sum(obj))
solver.Solve() # Solução 0 significa solução ótima do problema
```

```python
# Printando o resultado em forma de tabela
df = pd.DataFrame(columns=["X", "Preço", "Status"], index=np.arange(1, 31))

for j in J:
  df['X'][j]      = j
  df["Preço"][j]  = Lucro[j]
  df["Status"][j] = x[j].solution_value()

print(f"Valor Máximo da função objetivo: {solver.Objective().Value()}")
df
```
